# 2. Data Understanding

```python
# import library python for data science
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset

mobil = pd.read_csv('datasets/car_price.csv')
```

```python
# menampilkan tipe data masing-masing atribut
mobil.info()
```

# 3. Data Preparation

```python
# mengecek missing value
mobil.isnull().sum()
```

```python
# data reduction: menghilangkan atribut car_ID
mobil_new = mobil.drop(['car_ID'], axis=1)

# menampilkan 5 data teratas (tanpa atribut car_ID)
mobil_new.head()
```

```python
# melihat deskripsi sederhana dari data
mobil_new.describe()
```

```python
# menampilkan tabel correlation matrix
mobil_korelasi = mobil_new.corr(numeric_only=True)
mobil_korelasi
```

```python
# menampilkan correlation matrix dalam bentuk heatmapabs
plt.figure(figsize=(20,20))
sns.heatmap(mobil_korelasi, cbar=True, fmt='.2f', annot=True, annot_kws={'size':15}, cmap='YlGnBu')
plt.show()
```

```python
# atribut yang saling berhubungan erat dengan atribut price

sns.heatmap(mobil_korelasi[['price']].sort_values(by='price', ascending=False), annot=True, cmap='YlGnBu')
plt.show()
```

# 4. Modeling

```python
# data reduction
# ambil data pada atribut price (dependent/terikat)
price = mobil_new['price']

# ambil data pada atribut enginesize (independent/bebas)
enginesize = mobil_new['enginesize']
```

```python
price.head()
```

```python
enginesize.head()
```

```python
# karena masih dalam bentuk dataframe maka perlu diubah ke array
# mengubah dataframe menjadi data array

x = np.array(enginesize).reshape(-1,1)
y = np.array(price).reshape(-1,1)
x
```

```python
# menggunakan algoritme linear Regression dari sklearn
from sklearn.linear_model import LinearRegression
```

```python
# mencari nilai koefisien dan intercept
lr = LinearRegression()
lr.fit(x,y)

print('Nilai koefisien = ', lr.coef_)
print('Nilai intercept = ', lr.intercept_)
```

# 5. Evaluasi

```python
# memanggil library split data dan mse
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3, random_state=0)

# pemodelan dan evaluasi MSE
lr.fit(xtrain,ytrain)
ypred = lr.predict(xtest)


# mencari nilai MSE (Mean Square Error)

mse_lr = mean_squared_error(ytest,ypred)
print("Nilai MSE = ", mse_lr)
```

```python
# visualisasi Linear regression
plt.scatter(x,y)
plt.plot(x, lr.predict(x), color = 'red')
plt.show()
```
